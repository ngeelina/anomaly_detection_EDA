# Exploratory data analysis of credit card dataset

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### The dataset can be downloaded from https://www.kaggle.com/mlg-ulb/creditcardfraud link.

```python
credit_card_df = pd.read_csv('creditcard.csv') # read data from the csv file
credit_card_df.head()
```

#### The csv file contains 2,84,807 credit card transactions and has 31 columns

```python
credit_card_df.shape
```

#### Checking if there are any null values

```python
credit_card_df.isnull().values.any()
```

```python
credit_card_df.describe()
```

```python
df_corr = credit_card_df.corr()
plt.figure(figsize=(15,10))
sns.heatmap(df_corr, cmap="YlGnBu") # Displaying the Heatmap
sns.set(font_scale=2,style='white')

plt.title('Heatmap correlation')
plt.show()
```

#### It can be noted that most of the features are not correlated with each other.

```python
LABELS = ["Normal","Fraud"]
count_classes = pd.value_counts(credit_card_df['Class'], sort = True)
count_classes.plot(kind = 'bar', rot=0, color = 'blue')
plt.xticks(range(2), LABELS)
plt.title("Frequency by observation number")
plt.xlabel("Class")
plt.ylabel("Number of Observations");
```

#### Highly unbalanced dataset with 284315 normal cases and 492 fraudulent cases

```python
normal_df = credit_card_df[credit_card_df.Class == 0] #save normal_df observations into a separate df
fraud_df = credit_card_df[credit_card_df.Class == 1] #do the same for frauds
```

#### Statistics summary 

```python
normal_df.Amount.describe()
```

```python
fraud_df.Amount.describe()
```

#### Visualization of transaction data by time of the day, where time is measured from first transaction in the dataset. The dataset contains data for 2 days. It can be seen that frequency of normal trasactions decrease from hours 1 - 8 and again from hours 24 - 32. 
#### It can also be seen that frauds occur with higher frequency at this time. 

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12,6))
f.suptitle('Percentage of transactions by hour')

bins = np.linspace(0, 48, 48) #48 hours
ax1.hist((normal_df.Time/(60*60)), bins, alpha=1, density=True, label='Normal', color='green')
ax1.set_title('Normal')

ax2.hist((fraud_df.Time/(60*60)), bins, alpha=1, density=True, label='Normal', color='red')
ax2.set_title('Fraud')

plt.xlabel("Transaction time (hours)")
plt.ylabel("Percentage of transactions (%)")
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12,6))
f.suptitle('Time of transaction vs Amount by class')

ax1.scatter((normal_df.Time/(60*60)), normal_df.Amount, alpha=0.6, label='Normal', color='green')
ax1.set_title('Normal')

ax2.scatter((fraud_df.Time/(60*60)), fraud_df.Amount, alpha=0.9, label='Fraud', color='red')
ax2.set_title('Fraud')

plt.xlabel("Transaction time(hours)")
plt.ylabel('Amount')
plt.legend(loc='upper right')
plt.show()
```
